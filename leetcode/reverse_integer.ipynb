```python
'''
Level : Meduim



Given a signed 32-bit integer x, return x with its digits reversed. If reversing x causes the value to go outside the signed 32-bit integer range [-231, 231 - 1], then return 0.

Assume the environment does not allow you to store 64-bit integers (signed or unsigned).


Example 1:

Input: x = 123
Output: 321
Example 2:

Input: x = -123
Output: -321
Example 3:

Input: x = 120
Output: 21

'''
```

```python
### Answer ###
class Solution: 
    def reverse(self, x: int) -> int: 
        if x > 0: 
            # changing to str and perform reverse and then again change the type to integer : 
            x = int(str(x)[::-1])  
        else: 
            x = -1*int(str(x*-1)[::-1]) 
        if (-2**31 > x) or (x > 2**31 - 1): 
            return 0 
        else: 
            return x
```

```python
run = Solution()

print (run.reverse(123))
```

```python
print (run.reverse(-123))
```

```python
print (run.reverse(120))
```
